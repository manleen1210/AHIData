#Transformations

```python
def rules(row):
  if row['total_beds_7_day_avg'] >= 100:
    return 'large'
  elif row['total_beds_7_day_avg'] < 50:
    return 'small'
  else:
    return 'medium'
```

```python
df_COVIDFinal['total_beds_7_day_avgsize'] = df_COVIDFinal.apply(rules, 1)
```

```python
df_COVIDFinal[['total_beds_7_day_avg', 'total_beds_7_day_avgsize']]
```

```python
def rules2(row):
  if row['total_adult_patients_hospitalized_confirmed_covid_7_day_avg'] >= 25:
    return 'high_infection'
  elif row['total_adult_patients_hospitalized_confirmed_covid_7_day_avg'] < 25:
    return 'low_infection'
```

```python
df_COVIDFinal['Avg_infection_rate'] = df_COVIDFinal.apply(rules2, 1)
```
